## Session 2: Data Visualization and EDA

```python
# Importing Libraries
import numpy as np # For numerical computing
import pandas as pd # For data manipulation
import matplotlib.pyplot as plt # For data visualization
import seaborn as sns # For data visualization (better than matplotlib)
```

### Line Plot

```python
# Using Matplotlib

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot
plt.plot(x, y, label='Sine Wave', color='green', linestyle='--')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
# Using Seaborn

# Generate data
x = np.linspace(0, 10, 100)
y = np.cos(x)

# Create a line plot
sns.lineplot(x = x, y = y, label='Sine Wave', color='red', linestyle='solid')
plt.title('Line Plot')
plt.xlabel('XAxis')
plt.ylabel('YAxis')
plt.legend()
plt.show()
```

### Scatter Plot

```python
# Using Matplotlib

# Generate random data
np.random.seed(42)
x = np.random.rand(50)
y = 2 * x + 1 + 0.1 * np.random.randn(50)

# Create a scatter plot
plt.scatter(x, y, label='Data Points', color='blue', marker='o')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
# Using Seaborn

# Generate random data
np.random.seed(42)
x = np.random.rand(50)
y = 2 * x + 1 + 0.1 * np.random.randn(50)

# Create a scatter plot
sns.scatterplot(x = x, y = y, label='Data Points', color='green', marker='*')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

### Histogram 

<b>Definition:</b> A histogram is a graph that shows the frequency of numerical data using rectangles. The height of a rectangle (the vertical axis) represents the distribution frequency of a variable (the amount, or how often that variable appears).

<b>Note:</b> Histograms are used for numerical (continuous values) columns

```python
# Using Seaborn

# Generate random data
np.random.seed(42)
data = np.random.randn(1000)

# Create a histogram
sns.histplot(data, bins=30, kde=True, color='purple')
plt.title('Histogram with KDE')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# Using Matplotlib

# Generate random data
np.random.seed(42)
data = np.random.randn(1000)

# Create a histogram
plt.hist(data, bins = 50)
plt.title('Histogram with KDE')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

### Box Plot

<b>Box Plot:</b>In descriptive statistics, a box plot or boxplot is a method for graphically demonstrating the locality, spread and skewness groups of numerical data through their quartiles.
<br>
It is a standardized way of displaying the distribution of a data set based on its five-number summary of data points: the “minimum,” first quartile [Q1], median [Q2], third quartile [Q3] and “maximum.”

```python
# Using Seaborn

# Generate random data
np.random.seed(42)
data = np.random.randn(100, 4)

# Create a box plot
sns.boxplot(data=data, palette='Set3')
plt.title('Box Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# Using Matplotlib

# Generate random data
np.random.seed(42)
data = np.random.randn(1000)

# Create a box plot
plt.boxplot(data)
plt.show()
```

## Exploratory Data Analysis

Working on the IRIS dataset: The Iris dataset is a popular dataset in the field of machine learning and exploratory data analysis (EDA). It consists of 150 samples of iris flowers, each belonging to one of three species: setosa, versicolor, or virginica. The dataset has four features: sepal length, sepal width, petal length, and petal width, all measured in centimeters.

```python
from sklearn.datasets import load_iris # Importing the IRIS datset
```

```python
# Load Iris dataset
iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
data['species'] = iris.target_names[iris.target]
```

```python
## Univariate Analysis
# Univariate analysis - Histograms and KDE for each feature
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i+1)
    sns.histplot(data[feature], kde=True, bins=20)
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()
```

```python
# Bivariate analysis - Pairplot with regression lines
sns.pairplot(data, hue='species', height=2.5, kind='reg')
plt.show()
```

```python
# Multivariate analysis - Violin plot for feature distribution across species
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i+1)
    sns.violinplot(x='species', y=feature, data=data, inner='quartile')
    plt.title(f'Distribution of {feature} across Species')
plt.tight_layout()
plt.show()
```

```python
# Box plot for each feature across different species
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='species', y=feature, data=data)
    plt.title(f'Distribution of {feature} across Species')
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix heatmap
correlation_matrix = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
